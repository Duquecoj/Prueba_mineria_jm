```python

# Lista de variables meteorológicas
variables = ['fog', 'rain_drizzle', 'snow_ice_pellets', 'hail', 'thunder', 'tornado_funnel_cloud']

# 1. Convertir las columnas a numérico
for col in variables:
    if col in df.columns:
        # Convertir a string primero por si hay valores mixtos, luego a numérico
        df[col] = pd.to_numeric(df[col].astype(str), errors='coerce')
    else:
        print(f"Advertencia: Columna {col} no encontrada en el DataFrame")

# 2. Contar ocurrencias donde el valor es exactamente 1
conteos = {}
for var in variables:
    if var in df.columns:
        # Contar valores iguales a 1 (ignorando NaN)
        conteo = df[var].eq(1).sum()
        conteos[var] = int(conteo)  # Convertir a entero estándar

# 3. Mostrar resultados en formato claro
print("═"*60)
print(" RESUMEN DE OCURRENCIAS METEOROLÓGICAS ".center(60))
print("═"*60)
print(f"{'Evento':<25} | {'Total días':>10} | {'Porcentaje':>10}")
print("-"*60)

total_registros = len(df)
for evento, cantidad in conteos.items():
    porcentaje = (cantidad / total_registros) * 100 if total_registros > 0 else 0
    print(f"{evento.replace('_', ' ').title():<25} | {cantidad:>10,} | {porcentaje:>9.2f}%")

# 4. Crear DataFrame con resultados
df_resultados = pd.DataFrame.from_dict(conteos, orient='index', columns=['Días con evento'])
df_resultados['Porcentaje'] = (df_resultados['Días con evento'] / total_registros * 100).round(2)
print("\nTabla resumen:")
display(df_resultados.sort_values('Días con evento', ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
from matplotlib.ticker import MaxNLocator

# 1. Verificar y preparar los datos
if 'date' not in df.columns:
    print("Error: No se encuentra la columna 'date'")
else:
    try:
        # Convertir fecha considerando el formato dd/mm/YYYY
        df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y', errors='coerce')
        
        # Eliminar filas con fechas inválidas si las hay
        if df['date'].isna().any():
            print(f"Advertencia: Se eliminaron {df['date'].isna().sum()} filas con fechas inválidas")
            df = df.dropna(subset=['date'])
        
        # Extraer mes y nombre del mes
        df['month'] = df['date'].dt.month
        df['month_name'] = df['date'].dt.month.apply(lambda x: calendar.month_abbr[x])
        
        # 2. Procesar eventos climáticos
        eventos = ['fog', 'rain_drizzle', 'snow_ice_pellets', 'hail', 'thunder', 'tornado_funnel_cloud']
        eventos = [e for e in eventos if e in df.columns]  # Solo eventos existentes
        
        # Calcular ocurrencias por mes para cada evento
        datos_grafico = []
        for evento in eventos:
            temp_df = df[df[evento] == 1].groupby(['month', 'month_name']).size().reset_index(name='count')
            temp_df['evento'] = evento.replace('_', ' ').title()
            datos_grafico.append(temp_df)
        
        df_grafico = pd.concat(datos_grafico)
        
        # 3. Crear el gráfico
        plt.figure(figsize=(12, 7))
        sns.set_style("whitegrid")
        
        # Ordenar los meses cronológicamente
        month_order = [calendar.month_abbr[i] for i in range(1, 13)]
        df_grafico['month_name'] = pd.Categorical(df_grafico['month_name'], categories=month_order, ordered=True)
        
        # Gráfico de barras
        ax = sns.barplot(
            data=df_grafico,
            x='month_name',
            y='count',
            hue='evento',
            palette='viridis',
            order=month_order
        )
        
        # Personalización
        plt.title('Frecuencia Mensual de Eventos Climáticos', fontsize=16, pad=20)
        plt.xlabel('Mes', fontsize=12)
        plt.ylabel('Número de Días con Ocurrencia', fontsize=12)
        plt.legend(title='Evento Climático', bbox_to_anchor=(1.05, 1), loc='upper left')
        
        # Añadir valores a las barras
        for p in ax.patches:
            if p.get_height() > 0:
                ax.annotate(
                    f"{int(p.get_height())}",
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 5),
                    textcoords='offset points',
                    fontsize=9
                )
        
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Error al generar el gráfico: {str(e)}")
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

def get_nasa_power_data(lat, lon, start_date, end_date):
    """Obtiene datos de temperatura de la API NASA POWER"""
    base_url = "https://power.larc.nasa.gov/api/temporal/hourly/point"
    params = {
        'start': start_date,
        'end': end_date,
        'latitude': lat,
        'longitude': lon,
        'community': 'ag',
        'parameters': 'T2M',
        'header': 'true',
        'format': 'csv'
    }
    
    try:
        print(f"Obteniendo datos para lat={lat}, lon={lon}...")
        response = requests.get(base_url, params=params, timeout=30)
        response.raise_for_status()
        
        # Procesamiento especial del contenido
        content = response.text
        
        # Encontrar el inicio real de los datos CSV
        csv_start = 0
        for i, line in enumerate(content.split('\n')):
            if line.startswith("YEAR,MO,DY,HR"):
                csv_start = i
                break
                
        if csv_start == 0:
            raise ValueError("No se encontró el encabezado CSV en la respuesta")
            
        # Tomar solo las líneas de datos CSV
        csv_data = "\n".join(content.split('\n')[csv_start:])
        
        # Leer el CSV
        data = pd.read_csv(StringIO(csv_data))
        
        # Procesamiento de datos
        data['T2M'] = pd.to_numeric(data['T2M'], errors='coerce')
        data['datetime'] = pd.to_datetime(
            data['YEAR'].astype(str) + '-' +
            data['MO'].astype(str) + '-' +
            data['DY'].astype(str) + ' ' +
            data['HR'].astype(str) + ':00:00'
        )
        
        # Eliminar filas con valores nulos
        data = data.dropna(subset=['T2M'])
        
        return data[['datetime', 'T2M']].rename(columns={'T2M': 'temperature_2m'})
        
    except Exception as e:
        print(f"Error al obtener datos: {str(e)}")
        return None

# Uso de la función
temperature_data = get_nasa_power_data(
    lat=35,
    lon=-100,
    start_date='20240101',
    end_date='20241231'
)

if temperature_data is not None:
    print("\nDatos obtenidos exitosamente:")
    print(f"Registros: {len(temperature_data)}")
    print(f"Rango temporal: {temperature_data['datetime'].min()} a {temperature_data['datetime'].max()}")
    print(f"Temperatura promedio: {temperature_data['temperature_2m'].mean():.1f}°C")
    
    # Mostrar primeras filas
    display(temperature_data.head())
    
    # Gráfico
    plt.figure(figsize=(12, 6))
    plt.plot(temperature_data['datetime'], temperature_data['temperature_2m'], 
             color='tab:red', alpha=0.7, linewidth=0.5)
    plt.title('Temperatura a 2m de altura (2024)\nNASA POWER Data', fontsize=14)
    plt.xlabel('Fecha')
    plt.ylabel('Temperatura (°C)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Guardar datos
    temperature_data.to_csv('nasa_power_temperature.csv', index=False)
    print("\nDatos guardados en 'nasa_power_temperature.csv'")
else:
    print("No se pudieron obtener datos de la API")
```

```python
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, DateFormatter

# 1. Consulta optimizada para datos diarios en Celsius
query = """
SELECT 
    date as Fecha,
    AVG((temp-32)*5/9) as temp_celsius  # Conversión directa a Celsius en la consulta
FROM 
    `bigquery-public-data.noaa_gsod.gsod2024`
WHERE 
    date BETWEEN '2024-01-01' AND '2024-12-31'
    AND temp IS NOT NULL
    AND temp != 9999.9
GROUP BY 
    date
ORDER BY 
    date
"""

# 2. Obtener y procesar datos
client = bigquery.Client()
df = client.query(query).to_dataframe()
df['Fecha'] = pd.to_datetime(df['Fecha'])

# 3. Configurar gráfico
plt.figure(figsize=(14, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# 4. Gráfico de línea con datos crudos
plt.plot(df['Fecha'], df['temp_celsius'],
         color='#1E90FF',  # Azul estándar
         linewidth=1.5,
         alpha=0.8,
         label='Temperatura diaria')

# 5. Personalización
plt.title('Temperatura Promedio Diaria 2024 - NOAA GSOD', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)

# Configurar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Línea de promedio
promedio = df['temp_celsius'].mean()
plt.axhline(promedio, color='#FF6347', linestyle='--',
            label=f'Promedio: {promedio:.1f}°C')

plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

# 6. Mostrar resumen
print("\nResumen de datos:")
print(f"- Registros: {len(df)} días")
print(f"- Rango temporal: {df['Fecha'].min().date()} a {df['Fecha'].max().date()}")
print(f"- Temperatura promedio: {promedio:.1f}°C")
print(f"- Máxima: {df['temp_celsius'].max():.1f}°C")
print(f"- Mínima: {df['temp_celsius'].min():.1f}°C")

plt.show()

```

```python
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, DateFormatter

# 1. Función para obtener datos NOAA en Celsius
def get_noaa_data():
    query = """
    SELECT 
        date as Fecha,
        AVG((temp-32)*5/9) as temp_celsius  # Conversión directa a Celsius
    FROM 
        `bigquery-public-data.noaa_gsod.gsod2024`
    WHERE 
        date BETWEEN '2024-01-01' AND '2024-12-31'
        AND temp IS NOT NULL
        AND temp != 9999.9
    GROUP BY 
        date
    ORDER BY 
        date
    """
    client = bigquery.Client()
    df = client.query(query).to_dataframe()
    df['Fecha'] = pd.to_datetime(df['Fecha'])
    return df

# 2. Obtener datos NOAA
noaa_df = get_noaa_data()

# 3. Verificar datos NASA
if 'temperature_data' in globals() and temperature_data is not None:
    nasa_df = temperature_data.copy()
    nasa_df = nasa_df.rename(columns={'temperature_2m': 'temp_celsius'})
else:
    print("Datos NASA no disponibles")
    nasa_df = None

# 4. Crear gráfico comparativo
plt.figure(figsize=(14, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# Gráfico NOAA (azul)
plt.plot(noaa_df['Fecha'], noaa_df['temp_celsius'], 
         color='#1E90FF',  # Azul
         linewidth=1.5,
         alpha=0.8,
         label='NOAA GSOD')

# Gráfico NASA (rojo) si está disponible
if nasa_df is not None:
    plt.plot(nasa_df['datetime'], nasa_df['temp_celsius'],
             color='#FF6347',  # Rojo
             linewidth=0.9, 
             alpha=0.8, 
             label='NASA POWER')

# Configuración del gráfico
plt.title('Comparación de Temperaturas Diarias (2024)', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)

# Configurar eje X
ax = plt.gca()
ax.xaxis.set_major_locator(MonthLocator())
ax.xaxis.set_major_formatter(DateFormatter('%b %Y'))
plt.xticks(rotation=45)

# Añadir líneas de promedio
if not noaa_df.empty:
    noaa_avg = noaa_df['temp_celsius'].mean()
    plt.axhline(noaa_avg, color='#1E90FF', linestyle='--',
                label=f'Promedio NOAA: {noaa_avg:.1f}°C')

if nasa_df is not None:
    nasa_avg = nasa_df['temp_celsius'].mean()
    plt.axhline(nasa_avg, color='#FF6347', linestyle='--',
                label=f'Promedio NASA: {nasa_avg:.1f}°C')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Mostrar resumen comparativo
print("\nResumen Comparativo:")
if not noaa_df.empty:
    print(f"- NOAA GSOD: {len(noaa_df)} días | Promedio: {noaa_avg:.1f}°C")
if nasa_df is not None:
    print(f"- NASA POWER: {len(nasa_df)} registros | Promedio: {nasa_avg:.1f}°C")

plt.show()
```
